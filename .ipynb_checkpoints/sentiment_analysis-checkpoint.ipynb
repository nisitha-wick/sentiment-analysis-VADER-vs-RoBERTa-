## VADER Sentiment Scoring

```python
from nltk.sentiment import SentimentIntensityAnalyzer
from tqdm.notebook import tqdm

sia = SentimentIntensityAnalyzer()
```

```python
# running the polarity_score on the entire dataset
res = {}
for i, row in tqdm(df.iterrows(), total=len(df)):
    text = row['Text']
    myid = row['Id']
    res[myid] = sia.polarity_scores(text)
```

```python
vaders = pd.DataFrame(res).T
vaders = vaders.reset_index().rename(columns={'index' : 'Id'})
vaders = vaders.merge(df, how = 'left')
```

## RoBERTa Pretrained Model

```python
from transformers import TFAutoModelForSequenceClassification, AutoTokenizer
from scipy.special import softmax
```

```python
MODEL = f"cardiffnlp/twitter-roberta-base-sentiment"
tokenizer = AutoTokenizer.from_pretrained(MODEL)
model = TFAutoModelForSequenceClassification.from_pretrained(MODEL)
```

```python
# VADER results on example
print(example)
sia.polarity_scores(example)
```

```python
def polarity_scores_roberta(example):
    """
    RoBERTa cannot handle some sizes of the texts in the dataset so to handle
    truncation and padding ensures that the text fits within the model's accepatable input size.
    """
    encoded_text = tokenizer(example, return_tensors='tf',truncation=True, padding=True, max_length=512)
    output = model(**encoded_text)
    scores = output.logits.numpy()
    scores = softmax(scores, axis=-1)
    scores_dict = {
    'roberta_neg' : scores[0][0],
    'roberta_neu' : scores[0][1],
    'roberta_pos' : scores[0][2]
    }
    return scores_dict
```

```python
# running the polarity_score on the entire dataset
res = {}
for i, row in tqdm(df.iterrows(), total=len(df)):
    text = row['Text']
    myid = row['Id']
    vader_result = sia.polarity_scores(text)
    vader_result_rename = {}
    for key, value in vader_result.items():
        vader_result_rename[f"vader_{key}"] = value

    roberta_result = polarity_scores_roberta(text)

    both_results = {**vader_result_rename, **roberta_result}
    res[myid] = both_results
```

```python
result_df = pd.DataFrame(res).T
result_df = result_df.reset_index().rename(columns={'index' : 'Id'})
result_df = result_df.merge(df, how = 'left')
```
